# Analisi delle componenti connesse

Una componente connessa è un sottoinsieme di nodi nella rete che sono tutti interconnessi tra di loro
L'analisi delle componenti connesse nella rete aeroportuale mondiale può fornire informazioni importanti sulla struttura della rete e sulle sue proprietà

### Grafico risultante dal dataset

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

G = nx.read_graphml('openflights_world.graphml.gz')
fig, ax = plt.subplots(figsize=(25, 25))
nx.draw(G, node_size=1000 , with_labels=True, node_color="green", font_color="white", pos=nx.spring_layout(G))
plt.show()
```

### Quante componenti connesse ci sono?

```python
nx.number_connected_components(G)
```

Ci sono 7 componenti connesse, visibili anche dal grafico

### Quali sono?

```python
list(nx.connected_components(G))
```

### Componente connessa più grande

```python
core_nodes = max(nx.connected_components(G), key=len)
core = G.subgraph(core_nodes)
fig, ax = plt.subplots(figsize=(25, 25))
nx.draw_networkx(core, node_size=1000 , with_labels=True, node_color="red", font_color="white", pos=nx.spring_layout(G))
plt.show()
```

La rete non è diretta -> non ha senso chiedersi se è fortemente o debolmente connessa
